```python
import csv
file_name=[]
with open('TP_images.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        file_name.append(row[0])
    print(len(file_name))
```

```python
import os
path = './'
count=0
file=[]
for filename in os.listdir(path):
    #print(filename)
    if filename not in file_name:
        file.append(filename)
        count+=1
        #os.remove(filename)
print(count)
print(file)
```
